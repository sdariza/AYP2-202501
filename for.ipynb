````python
n = 0; prom = 0
while n <= 0:
  n = int(input("Ingrese el número de carros: "))
  if n <= 0:
    print("Número de carros inválido, por favor ingresa de nuevo el valor solicitado: ")
  else:
    print("Has ingresaso un número válido de carros")
for i in range(1, n + 1,1):
  print(f"Para el carro: {i}")
  p = float(input("Ingrese la presión: "))
  t = float(input("Ingrese la temperatura: "))
  v = float(input("Ingrese el volumen: "))
  prom = prom + (p * v) / 0.37 * (t + 460)
prom = prom / n
print(f"El promedio de la masa calculada para los {n} vehículos es: {prom}")
````
